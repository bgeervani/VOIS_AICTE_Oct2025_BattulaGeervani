```python
# Airbnb Open Data Analysis (Excel)
# pip install pandas matplotlib scikit-learn openpyxl
import pandas as pd
import matplotlib.pyplot as plt
# 1. Load dataset
df = pd.read_excel("/content/drive/MyDrive/1730285881-Airbnb_Open_Data.xlsx")
# 2. Basic info
print("Rows:", df.shape[0], " Columns:", df.shape[1])
print(df.columns)
# 3. Distribution of cancellation policies
df['cancellation_policy'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Distribution of Cancellation Policies")
plt.ylabel("Number of Listings")
plt.xlabel("Cancellation Policy")
plt.show()
# 4. Average price by room type
df.groupby('room type')['price'].mean().plot(kind='bar', color='orange')
plt.title("Average Price by Room Type")
plt.ylabel("Average Price")
plt.xlabel("Room Type")
plt.show()
# 5. Average availability by neighbourhood group
df.groupby('neighbourhood group')['availability 365'].mean().plot(kind='bar', color='green')
plt.title("Average Availability (days/year) by Neighbourhood Group")
plt.ylabel("Average Available Days")
plt.xlabel("Neighbourhood Group")
plt.show()
```
